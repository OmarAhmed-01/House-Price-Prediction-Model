```python
df["price"].fillna(value=df["price"].mean(),inplace=True)
df["bedrooms"].fillna(value=df["bedrooms"].mean(),inplace=True)
df["bathrooms"].fillna(value=df["bathrooms"].mean(),inplace=True)
df["sqft_living"].fillna(value=df["sqft_living"].mean(),inplace=True)
df["sqft_lot"].fillna(value=df["sqft_lot"].mean(),inplace=True)
df["floors"].fillna(value=df["floors"].mean(),inplace=True)
df["sqft_above"].fillna(value=df["sqft_above"].mean(),inplace=True)
df["sqft_basement"].fillna(value=df["sqft_basement"].mean(),inplace=True)
df["sqft_above"].fillna(value=df["sqft_above"].mean(),inplace=True)
df["sqft_living15"].fillna(value=df["sqft_living15"].mean(),inplace=True)
df["sqft_lot15"].fillna(value=df["sqft_lot15"].mean(),inplace=True)
df["grade"].fillna(value=df["grade"].mean(),inplace=True)
df["lat"].fillna(value=df["lat"].mean(),inplace=True)
df["long"].fillna(value=df["long"].mean(),inplace=True)
df["house_age"].fillna(value=df["house_age"].mean(),inplace=True)

#replace null values with the mean
```

```python
df['view']=df['view'].astype('category')
df['waterfront']=df['waterfront'].astype('category')
df['condition']=df['condition'].astype('category')

df.info()

#convert their DataTypes to category
```

```python
df_prep = pd.get_dummies(df,drop_first=True)

df_prep.head()
#creating dummy variables for category dtypes & checking the head
```

```python
X = df_prep.drop(['price'],axis=1)

Y = df_prep['price']

train_size=int(0.7*len(X))

#define X features and Y output & sets the train size 
```

```python
x_train = X.values[0:train_size]
x_test = X.values[train_size:len(X)]

y_train = Y.values[0:train_size]
y_test = Y.values[train_size:len(Y)]

#create train test sets
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
#create the linear regression model and create an object of it
```

```python
lr.fit(x_train,y_train)
#fit the data
```

```python
x_train
```

```python
y_predict=lr.predict(x_test)
#predict the price
```

```python
from sklearn import metrics 

MAE = metrics.mean_absolute_error(y_test,y_predict)
#mean absolute error
MSE = metrics.mean_squared_error(y_test,y_predict)
#mean squared error
RMSE = np.sqrt(metrics.mean_squared_error(y_test,y_predict))
#root mean squared error
R2 = lr.score(x_test,y_test)
#Coefficient of Determination 
print('Linear Regression Model \n')
print('MAE : {}\n'.format(MAE))
print('MSE : {}\n'.format(MSE))
print('RMSE : {}\n'.format(RMSE))
print('R2 : {}\n'.format(R2))
```

```python
sns.distplot((y_test-y_predict),bins=50);
#plotting the error histogram
```

```python
plt.figure(figsize=(8,8))
a = [1,16]
b =[1,16]

plt.scatter(y_test,y_predict,c='r',label = 'Actual Vs. Predicted')
plt.plot(a,b,c='b',label='45 deg line')
plt.xlabel('Actual Demand (log scale)')
plt.ylabel('Predicted Demand (log scale)')

plt.legend()
#plot the predicted vs. actual price with a 45 degrees line
```

```python
#Ridge Regression model
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
ridgeM=Ridge()
ridgeM.fit(x_train,y_train)
```

```python
y_pred_ridgeM=ridgeM.predict(x_test)
```

```python
Rridge=r2_score(y_test,y_pred_ridgeM)
Rrmse=np.sqrt(mean_squared_error(y_test,y_pred_ridgeM))
print('Ridge Regression Model \n')
print('RMSE : {}\n'.format(Rrmse))
print('R2 : {}\n'.format(Rridge))
```

```python
#Random Forest Regression
from sklearn.ensemble import RandomForestRegressor
RandomFM=RandomForestRegressor()
RandomFM.fit(x_train,y_train)
RandomFM.predictions=RandomFM.predict(x_test)
RFrmse=np.sqrt(mean_squared_error(y_test,RandomFM.predictions))
RFr2=r2_score(y_test,RandomFM.predictions)
print('Random Forest Regression\n')
print('RMSE : {}\n'.format(RFrmse))
print('R2 : {}\n'.format(RFr2))
```
